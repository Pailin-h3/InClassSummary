```python
class User:
    heros = []    # Hero
    rank = ''
    cash = 0
    coin = 0
    match_history = []    # Match
    friends = []    # User
    
    def __init__(self, name):
        self.name = name
    def __repr__(self):
        return 'User: %s\
                rank: %s\
                cash: %d\
                coin: %d' % (self.name,
                            self.rank,
                            self.cash,
                            self.coin)    # Can't loop Heros name
    def rankUP(self):
        pass 
    def rankDOWN(self):
        pass 
    def insert_cash(self, m):
        self.cash = self.cash + m
    def buyHero(self, c, hero, server):
        if(c == 'cash'):
            if(self.cash >= hero.cash_price):
                server.wallet = server.wallet + hero.cash_price
                self.cash = self.cash - hero.cash_price
                self.heros.append(hero)
            else:
                print("Your cash not enough")
        else:
            if(self.coin >= hero.coin_price):
                self.coin = self.coin - hero.coin_price
                self.heros.append(hero)
            else:
                print("Your coin not enough")

                
class Hero:
    skins = []    # Skin
    skills = []    # Skill
    cash_price = 0
    coin_price = 0
    
    def __init__(self, name, basic_skin, cash_price=0, coin_price=0):
        self.name = name
        self.skins.append(basic_skin)
        self.cash_price = cash_price
        self.coin_price = coin_price
    def buySkin(self, buyer, skin, server):
        if(self in buyer.heros):
            if(buyer.cash >= skin.price):
                server.wallet = server.wallet + skin.price
                buyer.cash = buyer.cash - skin.price
                self.skins.append(skin)
            else:
                print("cash not enough")
        else:
            print("Buy Hero First")
    
    
class Skin:
    pic = "SKIN SHOWN"
    price = 0
    
    def __init__(self, pic, price=0):
        self.pic = pic
        self.price = price

        
class Skill:
    str_damage = 0
    magic_damage = 0
    cooldown = 0
    
    def __init__(self, str_damage, magic_damage, cooldown):
        self.str_damage = str_damage
        self.magic_damage = magic_damage
        self.cooldown = cooldown

        
class InGameHero(Hero):
    skin = ''
    level = 0
    money = 0
    items = []    # Item
    useable_skills = []    # Skill
    total_dmg = 0
    kda = ''
    team = ''
    
    def __init__(self, skin='common'):
        self.skin = skin
    def buyItem(self, item):
        if(len(self.item) < 6):
            if(self.money >= item.buyprice):
                self.item.append(item)
                self.money = self.money - item.buyprice
            else:
                print("not enough money")
        else:
            print("Full ITEMS")
    def sellItem(self, idx):
        self.money = self.money + self.item[idx].sellprice
        self.item.remove(idx)
    
    
class InGameItem:
    buyprice = 0
    sellprice = 0
    property = ''
    components = []    # InGameItem    
    def __init__(self, bp, sp, ppt, *arg):
        self.buyprice = bp
        self.sellprice = sp
        self.property = ppt
        self.components.append(arg)
    
    
class Match(InGameItem, InGameHero):
    users = []    # User
    time = 0.00
    
    def dispense(self, coin, *args):
        for arg in args:
            arg.coin = arg.coin + coin

    
class Item:
    cash_price = 0
    coin_price = 0
    property = ''
    
    def __init__(self, property, cash_price = 0, coin_price = 0):
        self.property = property
        self.cash_price = cash_price
        self.coin_price = coin_price
    
    
class Shop:
    items = []    # Item
    skins = []    # Skin
    heros = []    # Hero

    
class Server:
    max_user = 0
    online_users = []    # User
    wallet = 0
```
