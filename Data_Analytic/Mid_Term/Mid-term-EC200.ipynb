# 1.2  (10 คะแนน)

ให้ นักศึกษาทำการเขียน code เพื่อ ทำการ คำนวนหาค่าสูงสุด (max) และค่าต่ำสุด (min) ของระดับ flavanoids ในไวน์ แต่ละประเภท และทำการ plot graph ด้วย Bar chart เพื่อแสดงระดับของสูงสุด (max) และค่าต่ำสุด (min) ในไวน์แต่ละประเภท ระบุค่าแกน x  แกน y และชื่อกราฟให้เหมาะสม พร้อมทำการวิเคราะห์สั้นๆว่าค้นพบอะไรจากข้อมูลนี้

```python
min_FLA = []
max_FLA = []
for i in range(len(target_names)):
    min_FLA.append(min(features[target==i,6]))
    max_FLA.append(max(features[target==i,6]))
print(min_FLA)
print(max_FLA)
```

```python
width = 0.35
x = np.arange(len(target_names))
fig, ax = plt.subplots()
rects1 = ax.bar(x-width/2,min_FLA,width,label='Min')
rects2 = ax.bar(x+width/2,max_FLA,width,label='Max')

ax.set_ylabel('flavanoids')
ax.set_title('max and min flavanoids of 3 wine')
ax.set_xticks(x)
ax.set_xticklabels(target_names)
ax.legend()

def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```

จากกราฟเห็นว่าสามารถใช้ flavanoid แยก class 2 ได้บ้างถ้าเกิน 8.9 ขึ้นไป แต่ส่วนใหญ่ยังปนๆกันอยู่

# 1.4  (10 คะแนน)

ทำการจำแนกประเภทของไวน์ class_0 และ class_1 โดยใช้ลักษณะของfeatures ทั้งหมด จากผลการจำแนก ได้ระดับ accuracy rate เท่ากับร้อยละเท่าไร ค่าจุดวิกฤติที่จะใช้ คืออะไร และ ทำจุดวิกฤติดังกล่าวมาแยกประเภทของไวน์อย่างไร จงอธิบายอย่างละเอียด

```python
labels = target_names[target]
not_class2 = (labels != 'class_2')
features2 = features[not_class2]
labels2 = labels[not_class2]
is_class0 = (labels2 == 'class_0')
```

```python
best_acc = -1.0

for fi in range(features2.shape[1]):
    thresh = features2[:,fi]
    
    for t in thresh:

        feature_i = features2[:, fi]
        pred = (feature_i > t)
        acc = (pred == is_class0).mean()
        rev_acc = (pred == ~is_class0).mean()
        
        if rev_acc > acc:
            reverse = True
            acc = rev_acc
        else:
            reverse = False

        if acc > best_acc:
            best_acc = acc
            best_fi = fi
            best_t = t
            best_reverse = reverse

print(best_fi, best_t, best_reverse, best_acc)
```

ระดับ accuracy rate เท่ากับร้อยละ 95.38
ค่าจุดวิกฤติที่จะใช้ คือ proline
ถ้า proline >= 750 คือ class 1
และ ถ้า proline < 750 คือ class 0

## 2.1 (15 คะแนน) 

กำหนดให้รูปแบบสมการข้อมูล มีลักษณะดังต่อไปนี้

$$
x_{t+1} =\theta+\alpha \, x_t + \epsilon_{t+1}
\quad \text{โดยที่} \quad
x_0 = 6
\quad \text{และ} \quad t = 0,\ldots,T
$$

ทั้งนี้เทอมของ $ \{\epsilon_t\} $ มีการกระจายตัวแบบ IID และมีการแจกแจงแบบปกติ (standard normal)
$$\{\epsilon_t\} \sim N(5,0.16)$$ หรือ mu =5 และ sigma=0.04 นั้นเอง

คำสั่ง:
ให้ทำการเขียน โค้ด โดยใช้คำสั่ง for ทำการสร้างชุดข้อมูล จำนวน 1000 ชุดข้อมูล โดยกำหนดให้ x_0=13 
$\theta$ =2 และ $\alpha$ =0.5
และทำการ plot กราฟดังกล่าว ระบุแกนและชื่อของกราฟ ให้เหมาะสม

คำใบ้ (Hint):
ใน python หากต้องการสร้างตัวแปรสุ่ม ที่มีการกระจายตัวแบบ N($\mu$,$\sigma^2$) เราสามารถเขียนโค้ดได้คือ

sigma * np.random.randn() + mu

```python
def x_function(α,x,θ,ϵ):
    return θ + α*x + ϵ

x0 = 6
α = 5
θ = 2
sigma = 0.04
mu = 5
ϵ = sigma * np.random.randn(1000) + mu
X = x_function(α,x0,θ,ϵ)
Y = np.arange(1000)
plt.plot(Y,X)
plt.xlabel('X value')
plt.ylabel('Time')
plt.show()
```

## 2.2 (15 คะแนน )

จากข้อมูลที่ได้ทำการสร้างขึ้นในข้อ 2.1 ทำการ plot histogram  และคำนวนหาค่าสถิติดังต่อไปนี้

2.2.1  Mean

2.2.3  Skewness

2.2.4  Kurtosis

จากการคำนวน นักศึกษา คิดว่า ค่า kurtosis ใกล้เคียงกับการกระจายตัวแบบ Normal distribution หรือ ไม่ เพราะเหตุใด

```python
meanx=sum(X)/len(X)
n=len(X)
sdx = (sum((X-meanx)**2)/(n-1))**0.5
skew_x = (sum((X-meanx)**3)/(n-1))/(sdx**3)
kurr_x = (sum((X-meanx)**4)/(n-1))/(sdx**4)
print('SD X = %s'% sdx)
print('Skewness X = %s'% skew_x)
print('Kurtosis X = %s'% kurr_x)
```

## 3.1  (10 คะแนน)
กำหนดให้
    $$\phi= r+3xy^2+x^3-50x$$
$$r=x^3+4y^3$$
จงเขียน definition ของค่า $\phi$ จากนั้น คำนวนหาค่า partial derivative ของ y with respect to x หรือเขียน เป็น สัญลักษณ์ ทางคณิตศาสตร์ได้คือ 
$$\frac{\partial \phi}{\partial x}$$

```python
from sympy import Symbol,diff
```

```python
def ϕ_function(x,y):
    return (x**3) + (4*y**3) + (3*x*y**2) + (x**3) - (50*x)

x = Symbol('x')
y = Symbol('y')
ϕ = ϕ_function(x,y)
diff(ϕ,x)
```

## 3.2  (10 คะแนน)
จากคำตอบที่ ได้ในข้อ 3.1 ทำการจัดรูปและเขียนใหม่ ให้อยู่ในลักษณะ y=f(x) 
จากนั้น กำหนดให้ ค่า x อยู่ในช่วง [-1,1] ทำการหาค่าของ y ที่สอดคล้องกับฟังก์ชัน f(x) จากนั้นทำการ plot graph (x,y)

```python
def y_function(x):
    return ((50-(6*x**2))/3)**(1/2)
x_value = np.arange(-1,1,0.00001)
y_value = y_function(x_value)
plt.plot(x_value,y_value)
plt.show()
```
