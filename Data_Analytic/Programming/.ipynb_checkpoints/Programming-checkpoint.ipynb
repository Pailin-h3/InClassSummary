```python
para1 = input()
ISBN = False
numcheck = 0
k = 10
for i in range(10):
    #print("numcheck = "+str(numcheck)+" + (" +para1[i]+ " * " +str(k)+ ")")
    numcheck += (int(para1[i])*k)
    k-=1
if(numcheck%11 == 0):
    print("Number is ISBN")
    ISBN = True
else:
    print("Number is NOT ISBN")
```

```python
para2 = input()
numcheck2 = 0
k2 = 10
k1 = 0
for i in range(9):
    #print("numcheck = "+str(numcheck)+" + (" +para1[i]+ " * " +str(k)+ ")")
    numcheck2 += (int(para2[i])*k2)
    k2-=1
while(numcheck2%11!=0):
    k1+=1
    numcheck2+=1
para2 = para2+str(k1)
print(para2)
```

```python
p1,p2,p3,n1,n2 = input().split()
won = 0
for i in range(int(n1),int(n2)+1):
    print("buy "+ str(i) + ": ")
    if(i == int(p1)):
        won += 10000
        print("got 10000")
    if(i%100 == int(p2)):
        won += 25
        print("got 25")
    if(i%1000 == int(p3)):
        won += 100
        print("got 100")
    print("--------------------------")
print(won)
```

```python
infile = open('C:/Users/Lenovo Demo/Desktop/Data_Analytic/Programming/answers.txt','r')
soln = infile.readline().strip()
outfile = open('C:/Users/Lenovo Demo/Desktop/Data_Analytic/Programming/out.txt','w')
for line in infile:
    l = line.strip().split()
    sid=l[0]
    ans = l[1:]
    A =''
    for a in ans:
        A +=a+' '
    while(len(A)<=25):
        A +=' '
    point = 0
    for k in range(len(soln)):
        if(A[k]==soln[k]):
            point += 1
    outfile.write(sid+', '+ str(point)+'\n')
infile.close()
outfile.close()
```
