## URLlib

```python
import urllib.request as ur
```

```python
print(ur)
```

```python
url = 'https://github.com/is381-1-2020/'
conn = ur.urlopen(url)
```

```python
conn.status
```

status 200 = OK

```python
data = conn.read() # ask for all data in web
```

```python
data[:200]
```

```python
for key, value in conn.getheaders(): # get only headers
    print(key, value)
```

```python
data # data is HTML, get HTML data first
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(data, 'html.parser')
soup # already beautiful??
```

```python
soup.title # can call by Tag
```

```python
soup.find_all(class_ = 'mr-3 mr-lg-0') # search by class
```

```python
for title in soup.find_all('li', {'class': 'mr-3 mr-lg-0'}): # filter tag and class
    print(title.text)
```
